```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
retriever = SentenceTransformer("all-MiniLM-L6-v2")  # Encodeur de documents

# Documents d'exemple
documents = [
    "L'apprentissage automatique est un sous-domaine de l'intelligence artificielle.",
    "Les réseaux de neurones sont largement utilisés dans le deep learning.",
    "La génération de texte peut être améliorée avec des méthodes de récupération d'informations."
]

# Encoder les documents
doc_embeddings = retriever.encode(documents)
doc_embeddings = np.array(doc_embeddings).astype('float32')

# Indexation avec FAISS
index = faiss.IndexFlatL2(doc_embeddings.shape[1])
index.add(doc_embeddings)
```

```python
def retrieve(query, top_k=2):
    query_embedding = retriever.encode([query]).astype('float32')
    distances, indices = index.search(query_embedding, top_k)
    retrieved_docs = [documents[idx] for idx in indices[0]]
    return retrieved_docs
```

```python
query = "Comment fonctionne le deep learning ?"
print(retrieve(query))
```

```python
def generate_with_rag(query):
    retrieved_docs = retrieve(query)
    context = "\n".join(retrieved_docs)
    prompt = f"Contexte : {context}\n\nQuestion : {query}\nRéponse :"

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    output = model.generate(**inputs, max_new_tokens=100)

    return tokenizer.decode(output[0], skip_special_tokens=True)

print(generate_with_rag("Qu'est-ce que le deep learning ?"))
```

```python
import torch.nn.functional as F
```

```python
def weighted_loss(logits, labels, weights):
    loss = F.cross_entropy(logits, labels, reduction="none")  # Perte individuelle par token
    weighted_loss = loss * weights  # Appliquer le poids
    return weighted_loss.mean()  # Moyenne pour obtenir une seule valeur de perte
```

```python
def compute_weights(tokenized_inputs):
    weights = torch.ones_like(tokenized_inputs["input_ids"], dtype=torch.float32)

    # Appliquer un poids plus élevé aux tokens du contexte récupéré
    for i, token in enumerate(tokenized_inputs["input_ids"][0]):
        if token in tokenizer.encode("Contexte :"):
            weights[0, i:] *= 2.0  # Doubler le poids pour le contexte
    return weights.to("cuda")
```

```python
def train_step(batch):
    model.train()

    tokenized_inputs = tokenizer(batch["text"], return_tensors="pt", padding=True, truncation=True).to("cuda")
    weights = compute_weights(tokenized_inputs)

    outputs = model(**tokenized_inputs, labels=tokenized_inputs["input_ids"])
    loss = weighted_loss(outputs.logits, tokenized_inputs["input_ids"], weights)

    loss.backward()  # Backpropagation
    optimizer.step()
    optimizer.zero_grad()

    return loss.item()
```

```python
# Tester la génération avec RAG
print(generate_with_rag("Explique-moi le machine learning."))

# Exemple de batch pour l'entraînement
batch = {"text": "Contexte : L'intelligence artificielle est une technologie en plein essor.\nQuestion : Qu'est-ce que l'IA ?\nRéponse : L'IA est une branche de l'informatique qui vise à créer des systèmes capables d'imiter l'intelligence humaine."}

# Lancer une étape d'entraînement
loss = train_step(batch)
print(f"Perte entraînement : {loss}")
```
